**The size of word tokens in Brown Corpus -**

```python
print('The size of word tokens in Brown Corpus - ', end = '')
len(brown.words())
```

**The word types in the Brown corpus -**

```python
texttypes = nltk.pos_tag(brown.words())
print('The words and their types are -')
texttypes
```

```python
types = []
for i in range(len(texttypes)):
  if(texttypes[i][1] not in types):
    types.append(texttypes[i][1])
print('The word types present in the Brown corpus are -')
types
```

```python
print('The total number of different word types are -',end=' ')
len(types)
```

**The tokens with most frequency -**

```python
words = brown.words()
freqlist = []
fdist = dict(nltk.FreqDist(w.lower() for w in words))
for k in fdist.keys():
  temp = []
  temp.append(k)
  temp.append(fdist[k])
  freqlist.append(temp)

descFreqList = sorted(freqlist, key=itemgetter(1), reverse=True)
print('The top 10 most frequent tokens in the Brown corpus are -')
descFreqList[:10]
```

```python
print('Frequency of each token in the brown corpus -')
descFreqList
```
